```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
from pandas_datareader import data as pdr
import seaborn as sns
import openpyxl
sns.set_style('whitegrid')
yf.pdr_override()
```

```python
#carteira com 5 ETF:
class Cart_5_etf:
    def __init__(self, etf_1,etf_2, etf_3, etf_4, etf_5, start, end):
        
        self.etf_1 = etf_1
        self.etf_2 = etf_2
        self.etf_3 = etf_3
        self.etf_4 = etf_4
        self.etf_5 = etf_5
        self.start = start
        self.end = end
       
        
    def pegar_cotacao(self):
        acao = [self.etf_1,self.etf_2, self.etf_3, etf_4, self.etf_5]
        download = pdr.get_data_yahoo(acao, start= self.start, end= self.end)
        return download['Adj Close']
    
    def calcular_retorno(self):
        acao = [self.etf_1, self.etf_2, self.etf_3, self.etf_4, self.etf_5]
        adj_close = pd.DataFrame()
        for i in acao:
            
            adj_close[i] = pdr.get_data_yahoo(i, start= self.start, end= self.end)['Adj Close']
        
        carteira_retorno= adj_close.diff() / adj_close.shift(1)
        retorno_medio = carteira_retorno.mean()
        carteira_retorno2 = adj_close.iloc[-1] / adj_close.iloc[0] - 1
        
        return carteira_retorno2
    def calcular_cov(self):
        
        acao = [self.etf_1, self.etf_2, self.etf_3, self.etf_4, self.etf_5]
        matrix_cov = pd.DataFrame()
        
        for i in acao:
            
            matrix_cov[i] = pdr.get_data_yahoo(i, start= self.start, end= self.end)['Adj Close']
        
        carteira_retorno= matrix_cov.diff() / matrix_cov.shift(1)
        cov = carteira_retorno.cov()
    
        return cov
    
    def calcular_corr(self):
        
        acao = [self.etf_1, self.etf_2, self.etf_3, self.etf_4, self.etf_5]
        matrix_corr = pd.DataFrame()
        
        for i in acao:
            
            matrix_corr[i] = pdr.get_data_yahoo(i, start= self.start, end= self.end)['Adj Close']
        
        carteira_retorno= matrix_corr.diff() / matrix_corr.shift(1)
        cov = carteira_retorno.corr()
    
        return cov
```
